```python
import pickle
import random
from neurocaps.extraction import TimeseriesExtractor
random.seed(123)
# dir = os.path.dirname(__file__)

# confounds = ["Cosine*", "aComp*", "Rot*"]
```

```python
parcel_approach = {"Schaefer": {"n_rois": 200, "yeo_networks": 7, "resolution_mm": 2}}

loop_extractor = TimeseriesExtractor(
    space="MNI152NLin6Asym", #  MNI152 2mm template 
    parcel_approach=parcel_approach,
    standardize="zscore_sample",
    use_confounds=False,
    detrend=True,
    low_pass=None,
    high_pass=None,
)
```

```python
bids_dir = f"/home/zachkaras/fmri/fmri_model_data/bids_formatted_clean"

loop_extractor.get_bold(
    bids_dir=bids_dir,
    pipeline_name='fmriprep-v24', # arbitrary name to align folder structure
    task="coding",
    condition="loop",
    condition_tr_shift=5,
    verbose=True,
    tr=0.8,
    progress_bar=True,  # Parameter available in versions >= 0.21.5
)
```

```python
# extractor.timeseries_to_pickle(output_dir="/home/zachkaras/fmri/fmri_model_data/midprocessing", filename="participant_timeseries.pkl")
with open("/home/zachkaras/fmri/fmri_model_data/midprocess/extracted_info_loops.pkl", 'wb') as f:
    pickle.dump(loop_extractor, f)
```
